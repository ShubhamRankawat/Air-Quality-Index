```python
from sklearn.preprocessing import Imputer
imputer = Imputer(missing_values = 'NaN',strategy = 'mean',axis = 0)
```

```python
imputer.fit(features.iloc[:,[0]])
features.iloc[:,[0]] = imputer.fit_transform(features.iloc[:,[0]])
```

```python
aqnew = pd.DataFrame(features)
aqnew.apply(lambda x: sum(x.isnull()),axis = 0) 
```

```python
X=aqnew.iloc[:,3]
y = aqnew.iloc[:,0]
```

```python
X
```

```python
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
X_train
```

```python
X_train=X_train.reshape(-1,1)
```

```python
X_train
```

```python
from sklearn import linear_model as lm
model = lm.LinearRegression()
results = model.fit(X_train,y_train)
```

```python
accuracy=model.score(X_train,y_train)
print('Accuracy of the model:',accuracy)
```

```python
# print coefficients
print('intercept:',model.intercept_)
print('slope:',model.coef_)
```

```python
X_test=X_test.reshape(-1,1)
```

```python
X_test
```

```python
# predictions from the model
predictions = model.predict(X_test)
print('ozone:',predictions,sep = '\n')
```

```python
plt.scatter(y_test,predictions)
```

```python
# evaluating the model
from sklearn.metrics import mean_squared_error,r2_score
X_train = X_train.reshape(-1,1)
y_train_prediction = model.predict(X_train)
X_test = X_test.reshape(-1,1)
y_test_prediction=model.predict(X_test)
```

```python
#model evaluation for training set
import numpy as np
rmse_training = (np.sqrt(mean_squared_error(y_train,y_train_prediction)))
r2_training = r2_score(y_train,y_train_prediction)
```

```python
print("the model performance for training set")
print("------------------------------------")
print('RMSE is {}'.format(rmse_training))
print('R2_score is {}'.format(r2_training))
```

```python
rmse_testing = (np.sqrt(mean_squared_error(y_test,y_test_prediction)))
r2_testing = r2_score(y_test,y_test_prediction)
```

```python
#model evaluation for testing set
import numpy as np
rmse_testing = (np.sqrt(mean_squared_error(y_test,y_test_prediction)))
r2_testing = r2_score(y_test,y_test_prediction)
```

```python
print("the model performance for testing set")
print("------------------------------------")
print('RMSE is {}'.format(rmse_testing))
print('R2_score is {}'.format(r2_testing))
```
